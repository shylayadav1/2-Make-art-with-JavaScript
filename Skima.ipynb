```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import numpy as np
from scipy.stats import pearsonr

# Load CSV data into a DataFrame
df = pd.read_csv('/content/drive/MyDrive/skima data/train.csv')
print(df.columns)
```

```python
jd_texts = df['jd_section'].astype(str).tolist()
resume_texts = df['resume_section'].astype(str).tolist()
section_names = df['years_of_exp'].tolist()
relevance_scores = df['relevance_score'].tolist()
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
column_names = df.columns
print(column_names)

#creating four lists as follows from relevant columbs of datafram
jd_text = df['jd_section'].tolist()
resume_text = df['resume_section'].tolist()
section_name = df['section_name'].tolist()
relevance_score = df['relevance_score'].tolist()


print("JD Text (first 10 elements):", jd_text[:10])
print("Resume Text (first 10 elements):", resume_text[:10])
print("Section Name (first 10 elements):", section_name[:10])
print("Relevance Score (first 10 elements):", relevance_score[:10])
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
jd_embeddings = model.encode(jd_text, convert_to_tensor=True)
resume_embeddings = model.encode(resume_text, convert_to_tensor=True)
```

```python
# Step 4: Compute the row-wise cosine similarity
cos_sim_scores = []
for jd_vec, resume_vec in zip(jd_embeddings, resume_embeddings):
    # Move tensors to CPU and convert to NumPy arrays before using cosine_similarity
    cos_sim = cosine_similarity(jd_vec.cpu().numpy().reshape(1, -1),
                               resume_vec.cpu().numpy().reshape(1, -1)).item()
    cos_sim_scores.append(cos_sim)

#.cpu(): We use the .cpu() method to move the tensors jd_vec and resume_vec from the GPU to the CPU.
#.numpy(): We then convert these CPU tensors to NumPy arrays using the .numpy() method.
#.reshape(1, -1): The cosine_similarity function expects 2D arrays. We use .reshape(1, -1) to ensure that our NumPy arrays have the shape (1, number of features).
```

```python
# Step 5: Store the cosine similarity scores back in the DataFrame
df['cos_sim_scores'] = cos_sim_scores

# Step 6: Save the DataFrame with the new column to a CSV file
df.to_csv('your_file_with_cosine_similarity.csv', index=False)
```

```python

# Create DataFrame with relevant data
result_df = pd.DataFrame({
    'section_name': section_names,
    'cos_sim': cos_sim_scores,
    'relevance_score': relevance_scores
})

# Calculate correlation grouped by section_name
correlation_results = result_df.groupby('section_name').apply(
    lambda group: pearsonr(group['cos_sim'], group['relevance_score'])[0]
)

# Print correlation results for each section
print("Correlation / Concordance for each section:")
print(correlation_results)
```
